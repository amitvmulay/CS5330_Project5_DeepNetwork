```python
import numpy as np
from tensorflow import keras
from tensorflow.keras import layers
from matplotlib import pyplot as plt
```

#Dataset - Fashion MNIST

All the experiments are performed on Fashion MNIST dataset


```python
np.random.seed(42)
num_classes = 10
input_shape = (28,28,1)

(x_train,y_train), (x_test, y_test) = keras.datasets.fashion_mnist.load_data()
x_train = x_train.astype("float32") / 255
x_test = x_test.astype("float32") / 255
plt.imshow(x_train[1])
x_train = np.expand_dims(x_train, -1)
x_test = np.expand_dims(x_test, -1)
print("x_train shape:", x_train.shape)
print(x_train.shape[0], "train samples")
print(x_test.shape[0], "test samples")


# convert class vectors to binary class matrices
y_train = keras.utils.to_categorical(y_train, num_classes)
y_test = keras.utils.to_categorical(y_test, num_classes)
```

```python
def getModel(filterCount = 32, kernelSize = 3, dropOut = 0.5, extra = False, poolSize = 2, activationFunc = "relu"):
  if (extra):
    model = keras.Sequential( 
      [
          keras.Input(shape=input_shape),
          layers.Conv2D(filterCount, kernel_size=(kernelSize, kernelSize), activation=activationFunc),
          layers.MaxPooling2D(pool_size=(2, 2)),
          layers.Conv2D(filterCount * 2, kernel_size=(kernelSize, kernelSize), activation=activationFunc),
          layers.MaxPooling2D(pool_size=(2, 2)),
          layers.Conv2D(filterCount * 4, kernel_size=(kernelSize, kernelSize), activation=activationFunc),
          layers.MaxPooling2D(pool_size=(2, 2)),
          layers.Flatten(),
          layers.Dropout(dropOut),
          layers.Dense(num_classes, activation="softmax"),
      ]
    )
  
  else:
    model = keras.Sequential( 
        [
            keras.Input(shape=input_shape),
            layers.Conv2D(filterCount, kernel_size=(kernelSize, kernelSize), activation=activationFunc),
            layers.MaxPooling2D(pool_size=(poolSize, poolSize)),
            layers.Conv2D(filterCount * 2, kernel_size=(kernelSize, kernelSize), activation=activationFunc),
            layers.MaxPooling2D(pool_size=(poolSize, poolSize)),
            layers.Flatten(),
            layers.Dropout(dropOut),
            layers.Dense(num_classes, activation="softmax"),
        ]
    )

  return model

def plotTrainTestAccuracy(history):
  # summarize history for accuracy
  plt.plot(history.history['accuracy'])
  plt.plot(history.history['val_accuracy'])
  plt.title('model accuracy')
  plt.ylabel('accuracy')
  plt.xlabel('epoch')
  plt.legend(['train', 'test'], loc='upper left')
  plt.show()
  # summarize history for loss
  plt.plot(history.history['loss'])
  plt.plot(history.history['val_loss'])
  plt.title('model loss')
  plt.ylabel('loss')
  plt.xlabel('epoch')
  plt.legend(['train', 'test'], loc='upper left')
  plt.show()
```

#Experiment: Number of Epochs with different activation functions

Train default Model

```python
# train and test of default model setting with different activation function
actAccArr = []
actFuncName =[]
def runExperiment(activationFunc = "relu"):

  model = getModel(activationFunc = "relu")
  model.summary()

  batch_size = 128
  epochs = 15

  model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])

  history = model.fit(x_train, y_train, batch_size=batch_size, epochs=epochs, validation_split=0.1)

  plotTrainTestAccuracy(history)

  score = model.evaluate(x_test, y_test, verbose=0)
  print("Test loss:", score[0])
  print("Test accuracy:", score[1])
  return score[1]

#Run experiment with activation function "relu"
actAccArr.append(runExperiment("relu"))
actFuncName.append("relu")
#Run experiment with activation function "tanh"
actAccArr.append(runExperiment("tanh"))
actFuncName.append("tanh")
#Run experiment with activation function "selu"
actAccArr.append(runExperiment("selu"))
actFuncName.append("selu")
#Run experiment with activation function "sigmoid"
actAccArr.append(runExperiment("sigmoid"))
actFuncName.append("sigmoid")
#Run experiment with activation function "softmax"
actAccArr.append(runExperiment("softmax"))
actFuncName.append("softmax")
#Run experiment with activation function "softplus"
actAccArr.append(runExperiment("softplus"))
actFuncName.append("softplus")
#Run experiment with activation function "softsign"
actAccArr.append(runExperiment("softsign"))
actFuncName.append("softsign")
#Run experiment with activation function "elu"
actAccArr.append(runExperiment("elu"))
actFuncName.append("elu")
#Run experiment with activation function "exponential"
actAccArr.append(runExperiment("exponential"))
actFuncName.append("exponential")

plt.plot(actFuncName,actAccArr)
plt.xlabel("Activation Functions")
plt.xlabel("Accuracy")
plt.xticks(rotation=45)
plt.title("Accuracy for activation function")
plt.show()
```

# Experiment : Filter count

Filter count for 1st Convolutional layer is set to 16, 32, 48, 64

Filter count for 2nd Convolutional layer is set to 32, 64, 96, 128

**Hypothesis:** With more number of convolutional filters prediction accuracy should increase.

**Result:** Prediction accuracy increased for more number of filters

```python
fCountArr = []
fAccuracyArr = []
for filterCount in range(16,65,16):
  model = getModel(filterCount = filterCount)
  model.summary()
  batch_size = 128
  epochs = 10

  model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])

  history = model.fit(x_train, y_train, batch_size=batch_size, epochs=epochs, validation_split=0.1)
  plotTrainTestAccuracy(history)
  score = model.evaluate(x_test, y_test, verbose=0)
  print("Test loss:", score[0])
  print("Test accuracy:", score[1])
  fCountArr.append(filterCount)
  fAccuracyArr.append(score[1])

plt.plot(fCountArr,fAccuracyArr)
plt.ylabel('Accuracy')
plt.xlabel('Filter Count')
plt.title('Accuracy for filter count')
plt.show()
```

# Experiment : Dropout rate

FOr this experiment dropout rate is varied from 0.2 to 0.8

**Hypothesis:** Test accuracy should be more for smaller dropout.

**Result:** Test accuracy is better for smaller dropout values.

```python
fSizeArr = []
fAccuracyArr = []
for dropout in np.arange(0.2,0.85,0.05):
  model = getModel(dropOut = dropout)
  batch_size = 128
  epochs = 10

  model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])

  history = model.fit(x_train, y_train, batch_size=batch_size, epochs=epochs, validation_split=0.1)
  #plotTrainTestAccuracy(history)
  score = model.evaluate(x_test, y_test, verbose=0)
  print("Dropout Rate:",dropout)
  print("Test loss:", score[0])
  print("Test accuracy:", score[1])
  fSizeArr.append(dropout)
  fAccuracyArr.append(score[1])

plt.plot(fSizeArr,fAccuracyArr)
plt.ylabel('Accuracy')
plt.xlabel('Dropout')
plt.title('Accuracy for Dropout')
plt.show()
```

# Experiment : Size of pooling layer filter

For this experiment pooling window size is is varied from 2 to 5

**Hypothesis:** Test accuracy should be dropped for larger pooling window size.

**Result:** Test accuracy is dropped for larger pooling window size.

```python
fSizeArr = []
fAccuracyArr = []
for poolsize in range(2,5):
  model = getModel(poolSize = poolsize)
  batch_size = 128
  epochs = 10

  model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])

  history = model.fit(x_train, y_train, batch_size=batch_size, epochs=epochs, validation_split=0.1)
  #plotTrainTestAccuracy(history)
  score = model.evaluate(x_test, y_test, verbose=0)
  print("Pooling window size:",poolsize, " x", poolsize)
  print("Test loss:", score[0])
  print("Test accuracy:", score[1])
  fSizeArr.append(poolsize)
  fAccuracyArr.append(score[1])

plt.plot(fSizeArr,fAccuracyArr)
plt.ylabel('Accuracy')
plt.xlabel('Pooling Window Size')
plt.title('Accuracy for Pool size')
plt.show()
```
